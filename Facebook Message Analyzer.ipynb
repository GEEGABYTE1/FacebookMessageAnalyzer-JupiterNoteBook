```python
import json

with open(path_to_file) as file:
    chat_history = json.load(file)
```

```python
import pandas as pd
```

```python
messages = pd.DataFrame(chat_history['messages'])
```

```python
messages
```

```python
def convert_time(timestamp):
    return pd.to_datetime(timestamp,unit='ms')
```

```python
convert_time(1538778675241)
```

```python
messages['date'] = messages['timestamp_ms'].apply(convert_time)
```

```python
messages 
```

```python
def get_month(date):
    return date.month

def get_year(date):
    return date.year 
```

```python
messages['month'] = messages['date'].apply(get_month)
messages['year'] = messages['date'].apply(get_year)
```

```python
messages
```

```python
import nltk 
from nltk.sentiment.vader import SentimentIntensityAnalyzer
```

```python
nltk.download('vader_lexicon')
```

```python
sentiment_analyzer = SentimentIntensityAnalyzer()
```

```python
sentiment_analyzer.polarity_scores('that\'s sweet')
```

```python
sentiment_analyzer.polarity_scores('sick')
```

```python
def get_polarity(text):
    return sentiment_analyzer.polarity_scores(text)['compound']
```

```python
messages['sentiment'] = messages['content'].apply(get_polarity)
```

```python
messages
```

```python
messages.groupby('sender_name').mean()
```

```python
year_month = messages.groupby(['month', 'year', 'sender_name']).mean().reset_index()
```

```python
year_month
```

```python
name_1 = year_month[year_month['sender_name'] == "Jane Austen"]['sentiment'].values
```

```python
name_1
```

```python
name_2 = year_month[year_month['sender_name'] == 'Lewis Carroll']['sentiment'].values
```

```python
name_2
```

```python
import matplotlib.pyplot as plt 
```

```python
plt.figure(figsize = (20, 10))
plt.plot(name_1, label="Jane")
plt.plot(name_2, label="Lewis")
plt.legend()
print('-'*6)
print("Here is the sentiment difference between {name1} and name {name2}".format(name1='Jane', name2='Lewis'))
print('-'*6)
```
